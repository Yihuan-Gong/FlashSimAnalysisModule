```python
import matplotlib.pyplot as plt
import yt
plt.style.use('default')

import sys
sys.path.append("/cluster/home/yhgong/")

from python.modules.features.plot.plot2d import *
```

```python
simFile = SimFileModel(
    simPath="/lfs/data/yhgong/productionRun/mergerAGN/MHD/Cooling_Feedback_ContiRandom15",
    hdf5FilePrefix="perseus_merger_hdf5_plt_cnt",
    fileSterMyr=1
)
```

```python
simName: str = "MergerAGN MHD jet15"
```

### Original Velocity Field 

```python
ds = yt.load(simFile.getHdf5Path(3000))
yt.SlicePlot(
    ds, 
    normal="z", 
    width=(100, "kpc"), 
    fields=("gas", "velocity_magnitude"),
).set_zlim(("gas", "velocity_magnitude"), 1e5, 1e9)
```

```python
fig1, ax1 = Plot2d().velocityFiltering(
    field=VelocityFilteringField.Vtotal,
    axis="z",
    simFile=simFile,
    calculationInfo=VelocityFilteringCalculationInfoModel(
        timeMyr=3000,
        rBoxKpc=50,
    ),
    plotInfo=Plot2dInfoModel(
        title=f"Vtotal {simName}",
        isLog=True,
        zlimMin=1e5,
        zlimMax=1e9
    )
)
```

### Los Velocity Dispersion
